```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
main_df=pd.read_csv("/content/drive/MyDrive/AnomalyDetectionFile.csv")
main_df.date=pd.to_datetime(main_df.date)
main_df=main_df.set_index(main_df.date)



```

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

li=["M"]
keys={"7D":"Week","M":"Month","120D":"Qtr"}




for y in li:
  # df5=main_df.groupby("product_id")
  # df6=df5.resample(y).agg({"sales":"sum","revenue":"sum","price":"mean","stock":"sum"})
  df7=main_df.groupby("store_id")
  df8=df7.resample(y).agg({"sales":"sum","revenue":"sum","stock":"sum"})
  # df6=df6.reset_index()
  df8=df8.reset_index()





```

```python
df_anomaly=pd.DataFrame(None)
#  In this for Loop We will Deal first with Store_ID based Anomalies
df=df8;
df=df.rename(columns={"date":y})
for i in df["store_id"].unique(): 
  c=0
  df1=pd.DataFrame(None)
  df2=pd.DataFrame(None)
  df3=pd.DataFrame(None)
  df1=df[df["store_id"]==i]
  df1[y]=df1[y].apply(lambda x: x.strftime('%Y-%m-%d'))
  Q1 = df1.quantile(0.25)
  Q3 = df1.quantile(0.75)
  IQR = Q3 - Q1
  df2=df1[(df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR)) ]
  df3=df2.isna()
  Total_list=[]
  for k in df3.columns:
    individual_list=[]
    a=-1
    for j in df3[k]:
      a=a+1
      if j==False:
        individual_list.append(df3.index[a])
    Total_list.append(individual_list)
  b=-1
  for p in df3.columns:
    b=b+1
    temp=pd.DataFrame(None)
    temp[keys[y]]=df1[y][Total_list[b]]
    temp["Greg_Month"]=pd.DatetimeIndex(temp[keys[y]]).month_name()
    temp["Greg_Year"]=pd.DatetimeIndex(temp[keys[y]]).year
    temp["store_id"]=i
    temp[p]=df1[p][Total_list[b]]
    if temp.empty==False:
      if(c==0):
        df_anomaly=df_anomaly.append(temp,ignore_index=True)    
        c=1
      
df_anomaly
                  

      
```

```python
df10=main_df.groupby("hierarchy2_id")
df11=df10.resample(y).agg({"sales":"sum","revenue":"sum","stock":"sum"})
df11=df11.reset_index()
```

```python
df_anomaly_reason_copy["H/L"]="N.A"
df_anomaly_reason_copy["% of deviation"]="N.A"
df_anomaly_reason_copy["value of deviation"]="N.A"

for i in df_anomaly_reason_copy.index:
  store_value=df_anomaly_reason_copy.iloc[i]["store_id"]
  para=df_anomaly_reason_copy.iloc[i]["parameter"]
  df_temp=df8[df8["store_id"]==store_value]
  mean_calc=df_temp[para].mean()
  ratio=df_anomaly_reason_copy.iloc[i][para]/mean_calc*100
  if df_anomaly_reason_copy.iloc[i][para] > mean_calc:
    df_anomaly_reason_copy.at[i,"H/L"]="H"
  else:
    df_anomaly_reason_copy.at[i,"H/L"]="L"
  df_anomaly_reason_copy.at[i,"% of deviation"]=abs(ratio-100)
  df_anomaly_reason_copy.at[i,"value of deviation"]=abs(mean_calc-df_anomaly_reason_copy.iloc[i][para])


df_anomaly_reason_copy
```
